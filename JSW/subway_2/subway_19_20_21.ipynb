```python
import numpy as np 
import pandas as pd 
```

```python
sw19 = pd.read_csv('files/subway_2019.csv', encoding='euc-kr')
sw20 = pd.read_csv('files/subway_2020.csv', encoding='euc-kr')
sw21 = pd.read_csv('files/subway_2021.csv', encoding='euc-kr')
dong_st = pd.read_csv('files/dong_station.csv', encoding='euc-kr')
```

```python
sw19
sw20
sw21
# dong_st
```

```python
dong_st.columns = ['지하철역', '호선명', '구', '동', '구_동']
dong_st
```

```python
# '지하철역'으로 데이터 병합
station_merge19 = pd.merge(sw19, dong_st, how='outer', on='지하철역')
station_merge20 = pd.merge(sw20, dong_st, how='outer', on='지하철역')
station_merge21 = pd.merge(sw21, dong_st, how='outer', on='지하철역')
station_merge19.dtypes

# 병합하면서 변환된 float 열을 int로 변환
# 병합된 데이터프레임 리스트
station_merges = [station_merge19, station_merge20, station_merge21]

for df in station_merges:
    for col in df.select_dtypes(include='float').columns:
        df[col] = df[col].fillna(0).astype(int)

station_merge19
station_merge20
station_merge21

# NaN 값 0으로 채우기 
station_merge19.fillna(0, inplace=True)
station_merge20.fillna(0, inplace=True)
station_merge21.fillna(0, inplace=True)

station_merge19
station_merge20
station_merge21

# 데이터프레임 리스트
station_data_list = [station_merge19, station_merge20, station_merge21]

# '지하철역'이 0이 아닌 데이터만 필터링
station_data_list = [df[df['지하철역'] != 0] for df in station_data_list]

# 결과 확인 (필요한 경우)
station_data19, station_data20, station_data21 = station_data_list
station_data_list

# 지하철역이 없는 데이터 제외 
# station_data19 = station_merge19[station_merge19['지하철역'] != 0]
# station_data20 = station_merge20[station_merge20['지하철역'] != 0]
# station_data21 = station_merge21[station_merge21['지하철역'] != 0]

station_data19
station_data20
station_data21

# # '지하철역'으로 병합한 csv파일 (확인차)
station_merge_all = pd.concat([station_data19 , station_data20 , station_data21])
station_merge_all.to_csv('station_merged_ex.csv', index=False, encoding='utf-8')
station_merge_all

# 동이 없는 데이터 제외 
station_merge_all = station_data19[station_data19['동'].notna()]
station_merge_all

# 필요없는 컬럼 삭제 
station_merge_all.drop('작업일자', axis=1, inplace=True)
station_merge_all
```

```python
# 총승하차인원 컬럼 생성 

# 데이터프레임 리스트
station_all = [station_merge19, station_merge20, station_merge21]

# 각 데이터프레임에 대해 합산 컬럼 추가
for df in station_all:
    df['총승하차인원'] = df[['유임승차인원', '무임승차인원', '유임하차인원', '무임하차인원']].sum(axis=1)

station_merge19
station_merge20
station_merge21
```
