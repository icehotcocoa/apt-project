데이터 병합

```python
import numpy as np 
import pandas as pd 
```

```python
# 두 데이터 불러오기
z_df = pd.read_csv('0.z.csv', encoding='utf-8-sig')
sb_df = pd.read_csv('3.sb.csv',encoding='utf-8-sig')
```

```python
# 데이터 확인 
z_df
sb_df
```

```python
merged_df = pd.merge(z_df, sb_df, on='행정동_ID', how='inner')
```

```python
merged_df
```

```python
merged_df.to_csv('merged_df.csv', index=False, encoding='utf-8-sig')
```
