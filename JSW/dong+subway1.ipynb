년도별 분기생성

```python
# 분기별 생성
def get_quarter(month):
    # 다음 코드 돌리고 오류 나서 문자열 변환 -> 실행 o 
    month = str(month).zfill(2)

    if month in ['01', '02', '03']:
        return '1-3월'
    elif month in ['04','05','06']:
        return '4-6월'
    elif month in ['07','08','09']:
        return '7-9월'
    else:
        return '10-12월'
    
# 분기별 컬럼 추가 
def add_quarter_columns(df):
    df['분기'] = df['월'].apply(get_quarter)
    return df 

year22 = add_quarter_columns(year22)
year23 = add_quarter_columns(year23)
year24 = add_quarter_columns(year24)
```

```python
print(year22)
print(year23)
print(year24)

# 확인차 저장 
# year24.to_csv('year24.csv')
```

```python
# 기존 '분기' 컬럼을 기준으로 1, 2, 3, 4분기로 변환하는 함수
def convert_to_quarter(quarter):
    if quarter == '1-3월':
        return 1
    elif quarter == '4-6월':
        return 2
    elif quarter == '7-9월':
        return 3
    elif quarter == '10-12월':
        return 4
    else:
        return None  # 예상하지 못한 값에 대해서 None 반환

# '분기' 컬럼을 변환하여 1, 2, 3, 4 분기로 저장
year22['분기2'] = year22['분기'].apply(convert_to_quarter)
year23['분기2'] = year23['분기'].apply(convert_to_quarter)
year24['분기2'] = year24['분기'].apply(convert_to_quarter)

# 결과 확인
year22
year23
year24
```

지하철 승객수의 합계

```python
# 2022, 2023, 2024년 데이터에 대해 분기별로 지하철 승객 수의 합계 계산
year22_sp_sum = year22.groupby(['연도', '분기', '행정동', '시군구'])['지하철_승객_수'].sum().reset_index()
year23_sp_sum = year23.groupby(['연도', '분기', '행정동', '시군구'])['지하철_승객_수'].sum().reset_index()
year24_sp_sum = year24.groupby(['연도', '분기', '행정동', '시군구'])['지하철_승객_수'].sum().reset_index()

# 결과 출력: 각 연도별 합계 계산 결과
print("2022년 지하철 승객 수 합계:")
print(year22_sp_sum)

print("2023년 지하철 승객 수 합계:")
print(year23_sp_sum)

print("2024년 지하철 승객 수 합계:")
print(year24_sp_sum)
```

지하철 승객수의 평균

```python
# 원래 값 : float으로 나옴  
# 2022, 2023, 2024년 데이터에 대해 분기별로 지하철 승객 수의 평균 계산
year22_mean = year22.groupby(['연도', '분기', '행정동', '시군구'])['지하철_승객_수'].mean().reset_index()
year23_mean = year23.groupby(['연도', '분기', '행정동', '시군구'])['지하철_승객_수'].mean().reset_index()
year24_mean = year24.groupby(['연도', '분기', '행정동', '시군구'])['지하철_승객_수'].mean().reset_index()

# 결과 출력: 각 연도별 평균 계산 결과
print("2022년 지하철 승객 수 평균:")
print(year22_mean)

print("2023년 지하철 승객 수 평균:")
print(year23_mean)

print("2024년 지하철 승객 수 평균:")
print(year24_mean)
```
