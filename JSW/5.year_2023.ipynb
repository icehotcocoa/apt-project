23년도 데이터 가공

```python
import numpy as np 
import pandas as pd
```

```python
year23 = pd.read_csv('data_2023.csv')
```

```python
year23
```

```python
# '기준_날짜'를 문자열로 변환하기
year23['기준_날짜'] = year23['기준_날짜'].astype(str) 
year23['기준_날짜'] = pd.to_datetime(year23['기준_날짜'],format='%Y%m%d').dt.strftime('%Y%m%d')
year23['기준_날짜']
```

```python
# YYYY-MM-DD 형식으로 변환
year23['기준_날짜'] = year23['기준_날짜'].astype(str)
year23['기준_날짜'] = pd.to_datetime(year23['기준_날짜'], format='%Y%m%d').dt.strftime('%Y년%m월%d일')

# 결과 확인
print(year23['기준_날짜'].head())
```
