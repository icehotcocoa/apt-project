```python
# 총_승객수 값이 0.0 제외 
year22_df = year22[year22['총_승객수'] != 0.0]
year22_df

year23_df = year23[year23['총_승객수'] != 0.0]
year23_df

year24_df = year24[year24['총_승객수'] != 0.0]
year24_df
```

```python
# 다른구 같은 동 확인
def same_dong(df, dong_name, gu_name):
    same_dong = df[df['행정동'] == dong_name] 
    other_gus = same_dong[same_dong['시군구'] !=gu_name] 
    unique_gus = other_gus[['시군구','행정동']].drop_duplicates().reset_index(drop=True)
    return unique_gus 

dong_name = "신사동"
gu_name = "강남구"

same_dong_in_other_gus = same_dong(year22_df, dong_name, gu_name)
print(f"'{dong_name}'이(가) '{gu_name}' 외에 속한 구 목록:")
print(same_dong_in_other_gus)
```

```python
# shinsadong = year22_df[year22_df['행정동'] == '신사동'] 

# print("신사동 목록:")
# print(shinsadong)

shinsadong = year22_df[year22_df['시군구'] == '강남구'] 

print("강남구 목록:")
print(shinsadong)
```

```python
# --분기별로 어느 행정동이 상위에 드는지 확인 (2022년)

# 그룹화 : 연도, 분기, 행정동 별로 총_승객수 합산
dong23 = year22_df.groupby(['연도', '분기', '행정동'])['총_승객수'].sum().reset_index()

# 2022년, 분기별로 총_승객수가 가장 많은 행정동 찾기
top_dong23 = dong23.loc[dong23.groupby(['연도','분기'])['총_승객수'].idxmax()]
top_dong23 = top_dong23.reset_index(drop=True)

print(top_dong23)


# --분기별로 어느 행정동이 상위에 드는지 확인 (2023년)

# 그룹화 : 연도, 분기, 행정동 별로 총_승객수 합산
dong23 = year23_df.groupby(['연도', '분기', '행정동'])['총_승객수'].sum().reset_index()

# 2022년, 분기별로 총_승객수가 가장 많은 행정동 찾기
top_dong23 = dong23.loc[dong23.groupby(['연도','분기'])['총_승객수'].idxmax()]
top_dong23 = top_dong23.reset_index(drop=True)

print(top_dong23)


# --분기별로 어느 행정동이 상위에 드는지 확인 (2024년)

# 그룹화 : 연도, 분기, 행정동 별로 총_승객수 합산
dong24 = year24_df.groupby(['연도', '분기', '행정동'])['총_승객수'].sum().reset_index()

# 2022년, 분기별로 총_승객수가 가장 많은 행정동 찾기
top_dong24 = dong24.loc[dong24.groupby(['연도','분기'])['총_승객수'].idxmax()]
top_dong24 = top_dong24.reset_index(drop=True)

print(top_dong24)
```
