```python
import numpy as np 
import pandas as pd
```

```python
sw_19 = pd.read_csv('file/station_merge19.csv')
sw_20 = pd.read_csv('file/station_merge20.csv')
sw_21 = pd.read_csv('file/station_merge21.csv')
```

```python
sw_19
sw_20
sw_21
```

```python
sw_19 = sw_19.drop(columns=['월','분기'])
sw_20 = sw_20.drop(columns=['월','분기'])
sw_21 = sw_21.drop(columns=['월','분기'])

sw_19 = sw_19.drop(columns=['사용월','구_동','작업일자'])
sw_20 = sw_20.drop(columns=['사용월','구_동','작업일자'])
sw_21 = sw_21.drop(columns=['사용월','구_동','작업일자'])

sw_19 = sw_19.drop(columns=['호선명_y','호선명_x'])
sw_20 = sw_20.drop(columns=['호선명_y','호선명_x'])
sw_21 = sw_21.drop(columns=['호선명_y','호선명_x'])

sw_19 = sw_19.drop(columns=['연도'])

sw_19
sw_20 
sw_21
```

```python
# '지하철역' 컬럼을 문자열로 변환
sw_19.loc[:, '지하철역'] = sw_19['지하철역'].astype(str)
sw_20.loc[:, '지하철역'] = sw_20['지하철역'].astype(str)
sw_21.loc[:, '지하철역'] = sw_21['지하철역'].astype(str)

# 결과 확인
print(sw_19.head())
print(sw_20.head())
print(sw_21.head())
```

```python
# '지하철역' 컬럼에서 고유값의 개수 확인
print(sw_19['지하철역'].nunique())
print(sw_20['지하철역'].nunique())
print(sw_21['지하철역'].nunique())
```

```python
# 수정된 DataFrame을 CSV로 저장
sw_19.to_csv('sw19.csv', index=False)
sw_20.to_csv('sw20.csv', index=False)
sw_21.to_csv('sw21.csv', index=False)
```
