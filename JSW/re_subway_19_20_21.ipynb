(2019년) 호선명, 지하철역, 시군구, 행정동, 년분기에 총승하차인원으로 groupby 해서 같은 데이터끼리 그룹화 후 승하차인원수 합산

```python
# '총승하차인원'을 기준으로 groupby 후 합산
group19_sum = sw_19_with_station.groupby(
    ['호선명', '지하철역', '시군구', '행정동', '년분기']
)['총승하차인원'].sum().reset_index()

# '총승하차인원'이 0인 데이터만 제외
group19_non_zero = sw_19_with_station[sw_19_with_station['총승하차인원'] != 0]

# 병합 예시 (groupby_sum 결과 병합)
merged = pd.merge(
    sw_19_with_station, 
    group19_sum[['호선명', '지하철역', '시군구', '행정동', '년분기', '총승하차인원']], 
    on=['호선명', '지하철역', '시군구', '행정동', '년분기'], 
    how='left', 
    suffixes=('', '_합산')
)

# 결과 확인
print(merged.head())  # 병합된 결과 출력
```

 행정동 ,지하철, 년분기별로 승하차인원 합산 groupby 

```python
# '행정동', '지하철역', '년분기'별로 '총승하차인원' 합산
grouped_data19 = merged.groupby(
    ['행정동', '지하철역', '년분기']
)['총승하차인원'].sum().reset_index()

# 결과 확인
print(grouped_data19)
```

합산 뒤, 총승하차인원 평균값 계산

```python
# '총승하차인원'이 0이 아닌 데이터로 필터링 (NaN 처리 후 평균 계산)
group19_non_zero = grouped_data19[grouped_data19['총승하차인원'] != 0]

# '총승하차인원' 의 평균 계산
group19_mean = grouped_data19.groupby(
    ['지하철역', '행정동', '년분기']
).agg(
    총승하차인원_평균=('총승하차인원', 'mean')  # 총승하차인원 평균 계산
).reset_index()

# 결과 확인
print(group19_mean)
```

결과 확인 후, 총승하차인원 평균값이 float이라 정수형 변환(보기편해서)

```python
# '총승하차인원_평균'을 정수형으로 변환
group19_mean['총승하차인원_평균'] = group19_mean['총승하차인원_평균'].astype(int)

# 결과 확인
print(group19_mean)
# group19_mean.to_csv('group19_mean.csv', index=False)
```
