```python
# 분기 변환

# 총_승객수가 0.0인것과 값이 있는것을 각각 년도별 데이터프레임 csv로 저장 
year_df = [year22, year23, year24]  # 연도별 데이터프레임 리스트
years = [2022, 2023, 2024]          # 해당 연도 리스트

# 분기 변환을 위한 매핑 딕셔너리
quarter_mapping = {'1-3월':'1분기',
                   '4-6월':'2분기',
                   '7-9월':'3분기',
                   '10-12월':'4분기',
                   '10-11월':'4분기'}

# 연도별 처리
for df, year in zip(year_df, years):
    # 총_승객수가 0.0인 행정동 목록
    zero_human = df[df['총_승객수'] == 0.0]
    zero_human['분기'] = zero_human['분기'].map(quarter_mapping)  # 분기 변환

    # 0.0인 행정동의 데이터
    zero_dong_sum = zero_human[['연도', '분기', '행정동','시군구', '총_승객수']]

    print(zero_dong_sum.isna().sum())

    # 총_승객수가 0.0이 아닌 행정동 목록
    is_human = df[df['총_승객수'] != 0.0]
    is_human['분기'] = is_human['분기'].map(quarter_mapping)  # 분기 변환

    # 0.0이 아닌 행정동의 데이터
    is_dong_sum = is_human[['연도', '분기', '행정동','시군구', '총_승객수',]]

    # CSV 파일로 저장
    zero_dong_sum.to_csv(f'year_{year}_zero_dong.csv', index=False)
    is_dong_sum.to_csv(f'year_{year}_is_dong.csv', index=False)

    # 출력 (작업 완료 메시지)
    print(f"\n{year}년도:")
    print(f"총_승객수가 0.0인 행정동 목록 저장: 'year_{year}_zero_dong.csv'")
    print(f"총_승객수가 0.0이 아닌 행정동 목록 저장: 'year_{year}_is_dong.csv'")
```
