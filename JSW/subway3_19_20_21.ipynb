```python
merged = pd.concat([sw19, sw20, sw21], axis=0, ignore_index=True)
merged
```

```python
merged = merged.rename(columns={'구': '시군구', '동': '행정동', '총승하차인원':'총_승객수'}, inplace=False)
```

```python
merged = merged.dropna(subset=['행정동'])
merged
```

```python
# 오류발생
```

merged 동별로 지하철역명, 갯수 확인

```python
# merged 데이터프레임의 컬럼 이름을 확인
print(merged.columns)

# '지하철역' 컬럼이 있는지 확인 후, 없으면 컬럼 이름을 변경
if '지하철역' not in merged.columns:
    # 예를 들어, '지하철역'이 '지하철역 이름'으로 되어 있다면, 이름을 수정
    merged.rename(columns={'지하철역 이름': '지하철역'}, inplace=True)

# '동'별로 지하철역을 그룹화하여 나열
grouped_by_dong = merged.groupby('행정동')['지하철역'].apply(list).reset_index()

# 0번 인덱스 행 삭제
grouped_by_dong = grouped_by_dong.drop(index=0)

# '지하철역' 컬럼에서 중복값을 제거하고 고유한 값만 남기기
grouped_by_dong['지하철역'] = grouped_by_dong['지하철역'].apply(lambda x: ', '.join(sorted(set(x))) if isinstance(x, list) else x)

# 결과 확인
print(grouped_by_dong.head())

# 수정된 DataFrame을 CSV로 저장
grouped_by_dong.to_csv('dong_sw.csv', index=False)
print('CSV 저장 성공')

# '지하철역' 컬럼에서 고유값의 개수를 확인
print(grouped_by_dong['지하철역'].nunique())
```

dong 에 지하철역 추가해서 확인 (22, 23, 24)

```python
# merged 데이터에서 지하철역과 행정동 병합

sw_dong = pd.merge(dong, merged[['지하철역','행정동']],on=['행정동'], how='left')
sw_dong
```

```python
sw_dong['지하철역'] = sw_dong['지하철역'].fillna(0)
sw_dong
```
