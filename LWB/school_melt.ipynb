```python
import pandas as pd
```

```python
s1 = pd.read_csv('school_merged.csv')
```

```python
s1.fillna(0, inplace=True)
```

```python
s1.info()
```

```python
s2_columns = [col for col in s1.columns if "school_count" in col or col == 'adm_nm']
s2 = s1[s2_columns]
s2.info()
```

```python
s2.rename(columns= 
          {'adm_nm':'행정동', 'school_count_191':'20191', 'school_count_232':'20233', 
           'school_count_192':'20193', 'school_count_201':'20201', 'school_count_202':'20203',
            'school_count_211': '20211','school_count_212':'20213','school_count_221':'20221', 
            'school_count_222':'20223', 'school_count_231':'20231'}, inplace=True)
```

```python
s3 = s2.melt(id_vars=['행정동']).copy()
s4 = s2.melt(id_vars=['행정동']).copy()
s3['variable'].unique()
```

```python
s4['varialbe2'] = s4['variable'].str[:-1] + s4['variable'].str[-1].replace({'1':'2', '3':'4'})
s4 = s4.drop(['variable'], axis=1)
print(s4)
```

```python
s4.columns = ['행정동', 'value', 'variable']
print(s3)
print(s4)
```

```python
merged_df = pd.merge(s3, s4, on='행정동', suffixes=('_x', '_y'))

# 'variable'과 'value' 컬럼을 하나의 컬럼으로 각각 분리하여 행을 확장
melted_variable = pd.melt(merged_df, id_vars=['행정동'], value_vars=['variable_x', 'variable_y'], var_name='variable_type', value_name='variable')
melted_value = pd.melt(merged_df, id_vars=['행정동'], value_vars=['value_x', 'value_y'], var_name='value_type', value_name='value')
```

```python
# '행정동' 기준으로 변수 및 값들을 연결
final_df = pd.DataFrame({
    '행정동': melted_variable['행정동'],
    '년분기': melted_variable['variable'],
    '학교수': melted_value['value']
})

final_df
```

```python
final = final_df.drop_duplicates()
final
```
