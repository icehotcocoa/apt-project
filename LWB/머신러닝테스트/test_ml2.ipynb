```python
def categorize_df(df):
    for category, biz_list in categories.items():
        if df in biz_list:
            return category
    return "기타"
```

```python
df['업종별카테고리'] = df['업종명'].apply(categorize_df)
df
```

```python
# 월매출과 년분기를 한 분기씩 미룸

def shift_within_group(group):
    group = group.sort_values('년분기')  
    group['년분기_미룸'] = group['년분기'].shift(-1)
    group['월매출_shift1'] = group['월매출(점포)'].shift(-1)
    return group
```

```python
test_data = grouped_df.groupby(['행정동', '업종별카테고리'], group_keys=False).apply(shift_within_group)
test_data
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
test_data['year'] = test_data['년분기'].astype(str).str[:4].astype(int)  # 연도 추출
test_data['quarter'] = test_data['년분기'].astype(str).str[4:6].astype(int)  # 분기 추출

# Train: 20191~20231(17), Test: 20232~20241(4)
train = test_data[(test_data['year'] >= 2019) & (test_data['year'] <= 2023) & (test_data['quarter'] <= 1)]
test = test_data[(test_data['year'] == 2023) & (test_data['quarter'] >= 2)]
```

```python
# Feature와 Target 분리
feature_data = ['년분기', '행정동', '인구수', '지역생활인구', '장기외국인', "주차장면적(면)", 
                '단기외국인', '주차장개수(개소)', '학교수', '학생수', '버스정류장수']
target_data = ['월매출_shift1']

X_train = train[feature_data]
y_train = train[target_data]
X_test = test[feature_data]
y_test = test[target_data]

# '행정동' 원-핫 인코딩
one_hot_encoded_dong_train = pd.get_dummies(X_train['행정동'], prefix='행정동')
one_hot_encoded_dong_test = pd.get_dummies(X_test['행정동'], prefix='행정동')

# 훈련 데이터와 테스트 데이터의 컬럼을 동일하게 맞춰줌
one_hot_encoded_dong_test = \
    one_hot_encoded_dong_test.reindex(columns=one_hot_encoded_dong_train.columns, fill_value=0)


# 숫자형 데이터 분리 및 결합
X_train_numeric = X_train.drop('행정동', axis=1)
X_test_numeric = X_test.drop('행정동', axis=1)

X_train_combined = pd.concat([X_train_numeric, one_hot_encoded_dong_train], axis=1)
X_test_combined = pd.concat([X_test_numeric, one_hot_encoded_dong_test], axis=1)
```

### Linear Regression

```python
# 데이터 표준화 (StandardScaler 사용)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_numeric)
X_test_scaled = scaler.transform(X_test_numeric)

# 원-핫 인코딩 데이터와 결합
X_train_combined = np.hstack([X_train_scaled, one_hot_encoded_dong_train.to_numpy()])
X_test_combined = np.hstack([X_test_scaled, one_hot_encoded_dong_test.to_numpy()])

# Linear Regression 모델 훈련 및 예측
model1 = LinearRegression()
model1.fit(X_train_combined, y_train)
y_pred = model1.predict(X_test_combined)
```

```python
# 평가
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

train_mse = mean_squared_error(y_train, model1.predict(X_train_combined))
test_mse = mean_squared_error(y_test, y_pred)
train_rmse = np.sqrt(train_mse)
test_rmse = np.sqrt(test_mse)
train_r2 = r2_score(y_train, model1.predict(X_train_combined))
test_r2 = r2_score(y_test, y_pred)
train_mae = mean_absolute_error(y_train, model1.predict(X_train_combined))
test_mae = mean_absolute_error(y_test, y_pred)

print(f"훈련 데이터 MAE: {train_mae:.2f}")
print(f"테스트 데이터 MAE: {test_mae:.2f}")
print(f"훈련 데이터 MSE: {train_mse:.2f}")
print(f"테스트 데이터 MSE: {test_mse:.2f}")
print(f"훈련 데이터 RMSE: {train_rmse:.2f}")
print(f"테스트 데이터 RMSE: {test_rmse:.2f}")
print(f"훈련 데이터 R²: {train_r2:.2f}")
print(f"테스트 데이터 R²: {test_r2:.2f}")

# 평균값 계산
train_mean = y_train.mean().values[0]  # Series에서 실제 값 추출
test_mean = y_test.mean().values[0]

print(f"훈련 데이터 평균값: {train_mean}")
print(f"테스트 데이터 평균값: {test_mean}")

# MAE 대비 비율 계산
train_mae = mean_absolute_error(y_train, model1.predict(X_train_combined))
test_mae = mean_absolute_error(y_test, y_pred)

train_mae_ratio = (train_mae / train_mean) * 100
test_mae_ratio = (test_mae / test_mean) * 100

print(f"훈련 데이터 MAE 비율: {train_mae_ratio:.2f}%")
print(f"테스트 데이터 MAE 비율: {test_mae_ratio:.2f}%")
```
