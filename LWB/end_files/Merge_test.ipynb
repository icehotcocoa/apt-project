```python
import pandas as pd
```

```python
sales = pd.read_csv(r'D:\Work\project\project2\oos\Merged_Sales_data_v2.1.csv')
parking = pd.read_csv('parking_end.csv')
school = pd.read_csv('school_end.csv')
bus = pd.read_csv('bus_counts_end.csv')
```

```python
def Data_Comparison(a, b, c):
    """
    a = 데이터A
    b = 데이터B
    c = 비교할컬럼
    """
    set_a = set(a[c])
    set_b = set(b[c])
    common_values = set_a & set_b  # A와 B에 모두 있는 값
    only_in_a = set_a - set_b  # A에만 있는 값
    only_in_b = set_b - set_a  # B에만 있는 값

    # 결과 출력
    print("A와 B에 모두 있는 값:")
    print(common_values)

    print("\nA에만 있는 값:")
    print(only_in_a)

    print("\nB에만 있는 값:")
    print(only_in_b)
```

```python
Data_Comparison(sales, parking, '행정동')
```

parking data 동 삭제 및 전처리

```python
# 행정동 컬럼에서 . 을 ·으로 변경
parking["행정동"] = parking["행정동"].str.replace(".", "·", )
```

```python
parking["행정동"] = parking["행정동"].str.replace("관악신사동", "관악구 신사동", )
parking["행정동"] = parking["행정동"].str.replace(r"\b항동\b", "오류2동", regex=True)
parking["행정동"] = parking["행정동"].str.replace("개포3동", "일원2동", )
parking["행정동"] = parking["행정동"].str.replace("상일1동", "상일동", )
parking["행정동"] = parking["행정동"].str.replace("상일2동", "강일동", )
```

```python
parking = parking[parking["행정동"] != ("은평구")]
parking = parking[parking["행정동"] != ("둔촌1동")]
parking = parking[parking["행정동"] != ("양천구")]
```

```python
Data_Comparison(sales, parking, '행정동')
```

school & bus 데이터 동 삭제 및 전처리

```python
school["행정동"] = school["행정동"].apply(lambda v: v if "관악구 신사동" in v else v.split()[-1])
bus["행정동"] = bus["행정동"].apply(lambda v: v if "관악구 신사동" in v else v.split()[-1])
```

```python
school["행정동"] = school["행정동"].str.replace("서울특별시 관악구 신사동", "관악구 신사동", )
school = school[school["행정동"] != ("둔촌1동")]
bus["행정동"] = bus["행정동"].str.replace("서울특별시 관악구 신사동", "관악구 신사동", )
bus = bus[bus["행정동"] != ("둔촌1동")]
```

```python
Data_Comparison(sales, school, '행정동')
```

```python
Data_Comparison(sales, bus, '행정동')
```
