```python
# 년분기 리스트
year_quarters = [20191, 20192, 20193, 20194, 20201, 20202, 20203, 20204, 20211,
                 20212, 20213, 20214, 20221, 20222, 20223, 20224, 20231, 20232,
                 20233, 20234, 20241, 20242]

# 년분기에서 2분기 뒤의 값을 매칭하는 함수
def get_target_year_quarter(current_quarter):
    try:
        current_index = year_quarters.index(current_quarter)
        target_index = current_index + 2
        if target_index < len(year_quarters):
            return year_quarters[target_index]
        else:
            return None  # 리스트 범위를 벗어난 경우 None 반환
    except ValueError:
        return None  # 리스트에 없는 값인 경우 None 반환

# 타겟년분기 컬럼 추가
# data['타겟년분기'] = data['년분기'].apply(get_target_year_quarter)
data['타겟년분기'] = data['년분기'].map(get_target_year_quarter)

# 결과 확인
data.info()
```

```python
def categorize_merged_data(merged_data):
    for category, biz_list in categories.items():
        if merged_data in biz_list:
            return category
    return "기타"
```

```python
data['업종별카테고리'] = data['업종명'].apply(categorize_merged_data)
merged_data = data
merged_data
```

```python
grouped_df = merged_data.groupby(['년분기', '행정동', '업종별카테고리']).agg({
    '월매출(점포)': 'sum',
    # '주중매출': 'sum',
    # '주말매출': 'sum',
    # '점포수': 'sum',
    '지역생활인구': 'mean',
    '장기외국인': 'mean',
    '단기외국인': 'mean',
    '주차장면적(면)': 'mean',
    '주차장개수(개소)': 'mean',
    '학교수': 'mean',
    '학생수': 'mean',
    '버스정류장수': 'mean'	
}).reset_index()

grouped_df
```

```python
# 월매출과 년분기를 한 분기씩 미룸

def shift_within_group(group):
    group = group.sort_values('년분기')  
    group['년분기_미룸'] = group['년분기'].shift(-2)
    group['월매출_shift2'] = group['월매출(점포)'].shift(-2)
    return group
```

```python
test_data = grouped_df.groupby(['행정동', '업종별카테고리'], group_keys=False).apply(shift_within_group)
test_data
```
