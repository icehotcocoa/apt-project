```python
def Data_Comparison(a, b, c):
    """a = 데이터A
       b = 데이터B
       c = 비교할컬럼
    """
    set_a = set(a[c])
    set_b = set(b[c])
    common_values = set_a & set_b  # A와 B에 모두 있는 값
    only_in_a = set_a - set_b  # A에만 있는 값
    only_in_b = set_b - set_a  # B에만 있는 값

    # 결과 출력
    print("A와 B에 모두 있는 값:")
    print(common_values)

    print("\nA에만 있는 값:")
    print(only_in_a)

    print("\nB에만 있는 값:")
    print(only_in_b)
```

```python
def process_data(data, year):

    #print(data.columns)

    data = data[["시간대구분", "행정동코드", "총생활인구수"]]
    data["년분기"] = year
    data = data[['년분기', '시간대구분', '행정동코드', '총생활인구수']]
    col = ['년분기', '시간', '행정동코드', '총생활인구']
    data.columns = col
    selected_local_data = data.groupby(["년분기", "행정동코드"])["총생활인구"].sum().round(0).reset_index()
    selected_local_data = selected_local_data.astype(int)
    print(selected_local_data.info())
    return selected_local_data
```

```python
import pandas as pd

# 문자열로 된 데이터프레임 이름 리스트
data_list = [
    "selected_local_20191", "selected_local_20192", "selected_local_20193", "selected_local_20194",
    "selected_local_20201", "selected_local_20202", "selected_local_20203", "selected_local_20204",
    "selected_local_20211", "selected_local_20212", "selected_local_20213", "selected_local_20214",
    "selected_local_20221", "selected_local_20222", "selected_local_20223", "selected_local_20224",
    "selected_local_20231", "selected_local_20232", "selected_local_20233", "selected_local_20234",
    "selected_local_20241", "selected_local_20242"
]

# globals()를 사용하여 데이터프레임 객체 리스트 생성
dataframes = [globals()[name] for name in data_list]


# 데이터프레임 병합
combined_df = pd.concat(dataframes, ignore_index=True)

# 결과 저장 (선택사항)
combined_df.to_csv('Living_Local_merged.csv', index=False)


# 결과 출력
print(combined_df["년분기"].unique())
```
