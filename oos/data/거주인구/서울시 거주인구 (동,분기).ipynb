동별 총 인구수 행 & 필요컬럼만 추출

```python
data3 = data2[data2['구분별'] == '계']
```

```python
# "동별" 컬럼에서 "합계"를 제외한 "??동" 데이터만 필터링
dong_only_data = data3[~data3['동별'].str.contains("합계|구$", na=False)]
gu_only_data = data3[~data3['동별'].str.contains("합계|동$", na=False)]

# 결과 확인
dong_only_data['동별'].unique
```

```python
dong_only_data.reset_index()
```

```python
# 고유 동의 개수 확인
unique_dong_count = dong_only_data['동별'].nunique()
print(f"고유 동의 개수: {unique_dong_count}")
```

```python
# 중복 값 확인
duplicate_check = dong_only_data['동별'].duplicated().any()
print(f"중복 값 존재 여부: {duplicate_check}")
```

```python
# 중복된 동 이름만 확인, 강남구, 관악구에 신사동 한개씩 존재
duplicate_dongs = dong_only_data[dong_only_data['동별'].duplicated(keep=False)]
print(duplicate_dongs['동별'])
```

분리된 동 인구수 합산

```python
dong_only_data.loc[dong_only_data['동별'] == '상일2동', '동별'] = '강일동'
dong_only_data.loc[dong_only_data['동별'] == '상일1동', '동별'] = '상일동'
dong_only_data.loc[dong_only_data['동별'] == '항동', '동별'] = '오류2동'
dong_only_data.loc[dong_only_data['동별'] == '개포3동', '동별'] = '일원2동'
dong_only_data['동별'] = dong_only_data['동별'].str.replace('.', '·', regex=False)
```

```python
# 특정 컬럼을 제거 (예: '컬럼명')
dong_only_data = dong_only_data.drop(columns=['구분별'])
```

```python
dong425 = dong_only_data.groupby(['동별']).sum().reset_index()
```

```python
dong425[dong425['동별'].str.contains('^상일|^오류|^강일', na=False)]
```

Float -> Int

```python
for d in dong425.columns[1:]:
    dong425[d] = dong425[d].astype(int)
```
