```python
import pandas as pd
```

```python
# 데이터 불러오기
dong = pd.read_csv(r'D:\Work\project\project2\oos\data\거주인구\Resident_Population_Dong423.csv')

d19 = pd.read_csv('Cleaned_Sales_data_2019.csv')
d20 = pd.read_csv('Cleaned_Sales_data_2020.csv')
d21 = pd.read_csv('Cleaned_Sales_data_2021.csv')
d22 = pd.read_csv('Cleaned_Sales_data_2022.csv')
d2324 = pd.read_csv('Cleaned_Sales_data_2023,2024.csv')
d19.info()
```

```python
set_a = set(d19['행정동'])
set_b = set(dong['행정동'])
common_values = set_a & set_b  # A와 B에 모두 있는 값
only_in_a = set_a - set_b  # A에만 있는 값
only_in_b = set_b - set_a  # B에만 있는 값

# 결과 출력
print("A와 B에 모두 있는 값:")
print(common_values)

print("\nA에만 있는 값:")
print(only_in_a)

print("\nB에만 있는 값:")
print(only_in_b)
```

```python
def check_data(data):
    year = str(data['년분기'][0])[:4]
    print('{0}년 데이터 확인'.format(year))
    print('년분기 갯수 : ', data['년분기'].nunique())
    print('업종명 갯수 : ', data['업종명'].nunique())
    print('행정동 갯수 : ', data['행정동'].nunique(), '\n')
```

```python
check_data(d19)
check_data(d20)
check_data(d21)
check_data(d22)
check_data(d2324)
```
